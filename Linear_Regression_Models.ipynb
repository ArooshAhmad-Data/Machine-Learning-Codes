using the function **make_regression**  you can create a custom (randomly initialized) regression dataset

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X, Y = make_regression(n_features= 1, noise = 10, n_samples = 1000)
```

```python
# See the results of dataset made using make_regression()
plt.xlabel("Feature -X")
plt.ylabel("Feature -Y")
plt.scatter(X,Y, s=5)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# the fit function fit's the model i.e. trains on the model
lr.fit(X,Y)
```

Gives the coefficients og the equation
** y = w0 + w1x **
where w1 is the coefficient

```python
lr.coef_
```

```python
lr.intercept_ # the intexcept w0
```


#### remember that input should be a 2D array

if you write **lr.predict(2)** or **lr.predict([2])**, it will resultin an error

```python
lr.predict([[2]])
```

```python
x_pred = lr.predict(X)
```

```python
plt.scatter(X,Y, s=5, label='training')
plt.scatter(X, x_pred, s=5 , label='prediction')
plt.legend() # displays the label of the plot.
plt.xlabel("Feature - X")
plt.ylabel("Target - Y")
plt.show()
```
